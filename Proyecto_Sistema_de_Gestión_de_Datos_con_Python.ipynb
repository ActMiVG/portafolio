# **PROYECTO: SISTEMA DE GESTIÓN DE DATOS CON PYTHON**

Una vez visto el ***Módulo 1: Fundamentos de Programación***, se presenta la siguiente actividad para que el alumnado pueda poner a prueba lo aprendido durante este módulo.

Dentro de la siguiente dinámica, realizar un **menú CRUD** donde el tema a emplear es libre. Cada función deberá ejecutar una **operación del CRUD**:

*   **Ver todos los registros**
*   **Crear nuevo registro**
*   **Consultar registro**
*   **Actualizar registro**
*   **Eliminar registro**
*   **Salir menú**

```python
# El Sistema de Gestión de Datos almacenará las quinielas del mundial para los partidos de México en la fase de grupos
# Crear menú
menu = """
***********************************
              Menú
***********************************
1) Ver todos los registros
2) Crear nuevo registro
3) Consultar registro
4) Actualizar registro
5) Eliminar registro
6) Salir menú
************************************
"""

# Crear programación de partidos
programacion = """
---------------------------------------
|  Copa Mundial de Fútbol · Grupo A
---------------------------------------
|  México vs Sudáfrica
|  11/6/26
|
|  México vs Corea del Sur
|  18/6/26
|
|  México vs Por definir
|  24/6/26
---------------------------------------
"""

# Crear lista de partidos
partidos_grupo_a = [["México","Sudáfrica"], ["México","Corea del Sur"], ["México","Por definir"]]

# Crear estructura de datos: quiniela para el grupo de México en el mundial 2026
quinielas = {}

quinielas[1] = ["Misael", {('México', 'Sudáfrica'): [2, 0], ('México', 'Corea del Sur'): [3, 2], ('México', 'Por definir'): [1, 1]}]
quinielas[2] = ["Pedro", {('México', 'Sudáfrica'): [4, 1], ('México', 'Corea del Sur'): [0, 2], ('México', 'Por definir'): [0, 1]}]
```

```python
# función para ver los registros
def verQuiniela():
  if len(quinielas) == 0:
    print("No hay registros")
  else:
    for posicion, quiniela in quinielas.items():
      print(f"\nID: {posicion}\n Participante: {quiniela[0]}")
      print(f" {partidos_grupo_a[1][0]} {quiniela[1]['México', 'Sudáfrica'][0]}-{quiniela[1]['México', 'Sudáfrica'][1]} {partidos_grupo_a[1][1]}")
      print(f" {partidos_grupo_a[1][0]} {quiniela[1]['México', 'Corea del Sur'][0]}-{quiniela[1]['México', 'Corea del Sur'][1]} {partidos_grupo_a[1][1]}")
      print(f" {partidos_grupo_a[1][0]} {quiniela[1]['México', 'Por definir'][0]}-{quiniela[1]['México', 'Por definir'][1]} {partidos_grupo_a[1][1]}")

verQuiniela()
```

```python
# función para agregar quiniela
def crearQuiniela():
  resultados = {}

  for partido in partidos_grupo_a:
    print(f"\nPartido: {partido[0]} vs {partido[1]}")
    goles_local = int(input(f"Goles de {partido[0]}: "))
    goles_visitante = int(input(f"Goles de {partido[1]}: "))

    resultados[tuple(partido)] = [goles_local, goles_visitante]
  return resultados

#crearQuiniela()

# función para guardar quiniela
def guardarQuiniela(quiniela, nombre):

  if len(quinielas) > 0:
    posicion = max(quinielas.keys()) + 1
  else:
    posicion = 1

  quinielas[posicion] = [nombre, quiniela]
  print("\nQuiniela guardada\n\n")

#guardar_quiniela(resultados, "Misael")


```

```python
# función para consultar quiniela

def consultarQuiniela():
  encontrado = False

  nombre = input("Ingrese el nombre del participante: ")
  for posicion, quiniela in quinielas.items():
    if quiniela[0].lower() == nombre.lower():
      print(f"\n Participante: {quiniela[0]}")
      print(f" {partidos_grupo_a[1][0]} {quiniela[1]['México', 'Sudáfrica'][0]}-{quiniela[1]['México', 'Sudáfrica'][1]} {partidos_grupo_a[1][1]}")
      print(f" {partidos_grupo_a[1][0]} {quiniela[1]['México', 'Corea del Sur'][0]}-{quiniela[1]['México', 'Corea del Sur'][1]} {partidos_grupo_a[1][1]}")
      print(f" {partidos_grupo_a[1][0]} {quiniela[1]['México', 'Por definir'][0]}-{quiniela[1]['México', 'Por definir'][1]} {partidos_grupo_a[1][1]}")
      encontrado = True
      break
  if encontrado == False:
    print("\nQuiniela no encontrada")

#consultarQuiniela()
```

```python
# Crear función para actualizar registro
def actualizarQuiniela():
  actualizado = False

  nombre = input("\nIngrese el nombre del participante al cual actualizar su quiniela: ")
  for posicion, quiniela in quinielas.items():
    if quiniela[0].lower() == nombre.lower():
      print(f"\nQuiniela actual:\n {quiniela[1]}")

      # Solicitar actualización
      print("\n Registra la actualización de tu quiniela")
      actualizacion = crearQuiniela()
      quinielas[posicion] = [nombre, actualizacion]

      actualizado = True
      break

  if actualizado == False:
    print(f"\nQuiniela no registrada para {nombre}, prueba con otra")
```

```python
# Crear función para eliminar quiniela
def eliminarQuiniela():
  encontrado = False

  nombre = input("Ingrese el nombre del participante a eliminar: ")
  for posicion, quiniela in quinielas.items():
    if quiniela[0].lower() == nombre.lower():
      print("\nRegistro eliminado:")
      print(f"Nombre: {quiniela[0]} | Quiniela: {quiniela[1]}")
      # Eliminar coincidencia
      quinielas.pop(posicion)
      encontrado = True
      break

  if encontrado == False:
    print(f"\nQuiniela no registrada para {nombre}, prueba eliminar otra")
```

```python
# Ejecución del Sistema de Gestión de Datos
while True:
  print(menu)
  opcion = int(input("¿Qué deseas hacer?: "))

  if opcion == 1:
    verQuiniela()
  elif opcion == 2:
    nombre = input("\n¿Cuál es tu nombre?: ")
    res = crearQuiniela()
    guardarQuiniela(quiniela = res, nombre = nombre)
  elif opcion == 3:
    consultarQuiniela()
  elif opcion == 4:
    actualizarQuiniela()
  elif opcion == 5:
    eliminarQuiniela()
  elif opcion == 6:
    break
  else:
    print("Opción no válida")
```
